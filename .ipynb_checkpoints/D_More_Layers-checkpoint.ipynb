```python
X=df.drop('Strength',axis=1).values
y=df['Strength'].values
```

# C. Model With 100 Epoch

```python
X_norm = (X - X.mean()) / X.std()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm, 
                                                    y, 
                                                    test_size=0.3,
                                                    random_state=101)
```

```python
mse_list = []

for i in range(50):
    model= Sequential()

    model.add(Dense(10, activation="relu",input_shape=(X.shape[1],)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer='adam',
                  loss='mean_squared_error')
    
    model.fit(x=X_train,y=y_train, epochs=100,verbose=0)
    
    y_pred = model.predict(X_test)
    
    mse_list.append(mean_squared_error(y_test, y_pred))
    
```

```python
print("Mean of Mean Squared Error:", np.mean(mse_list))
print("Standard Deviation of Mean Squared Errors:", np.std(mse_list))
```

# Results

<font color='red'> After we normalized data, increased the epochs and number of hidden layers, the mean squared error has decreased.
